# Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Data Read-in

```python
df = pd.read_csv('data/Xeek_Well_15-9-15.csv')
df['DPHI'] = (2.65 - df['RHOB'])/1.65
```

```python
df.describe()
```

# Basic Plot

```python
plt.figure(figsize=(10, 8))
plt.scatter(df["NPHI"], df["DPHI"]);
```

```python
# Just for comparison
# df.plot.scatter("NPHI", "DPHI", figsize=(10, 8));
```

# Adding color with a 3rd variable

```python
plt.figure(figsize=(10,8))
plt.scatter(df['NPHI'], df['DPHI'], c=df['GR'], vmin=0, vmax=100, cmap='viridis_r')
plt.xlim(0, 0.8)
plt.ylim(0, 0.8)
plt.colorbar();
```

# Remove spines from the plot

```python
plt.figure(figsize=(10,8))
plt.scatter(df['NPHI'], df['DPHI'], c=df['GR'], vmin=0, vmax=100, cmap='viridis_r')
plt.xlim(0, 0.8)
plt.ylim(0, 0.8)
plt.colorbar()

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
```

# Adding titles and axes labels

```python
plt.figure(figsize=(10,8))
plt.scatter(df['NPHI'], df['DPHI'], c=df['GR'], vmin=0, vmax=100, cmap='viridis_r')
plt.xlim(0, 0.8)
plt.ylim(0, 0.8)
plt.colorbar(label='Gamma Ray (API)')

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.title('Density Porosity vs Neutron Porosity Scatter Plot', fontsize=14, fontweight='bold')
plt.xlabel('Neutron Porosity (dec)')
plt.ylabel('Density Porosity (dec)')
plt.show()
```

# Adding a faint grid

```python
plt.figure(figsize=(10,8))
plt.scatter(df['NPHI'], df['DPHI'], c=df['GR'], vmin=0, vmax=100, cmap='viridis_r')
plt.xlim(0, 0.8)
plt.ylim(0, 0.8)
plt.colorbar(label='Gamma Ray (API)')

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.title('Density Porosity vs Neutron Porosity Scatter Plot', fontsize=14, fontweight='bold')
plt.xlabel('Neutron Porosity (dec)')
plt.ylabel('Density Porosity (dec)')

plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.grid(color='lightgray', alpha=0.5, zorder=1)  # Use of zorder

plt.scatter(df['NPHI'], df['DPHI'], c=df['GR'], vmin=0, vmax=100, cmap='viridis_r', zorder=2)  # Use of zorder
plt.xlim(0, 0.8)
plt.ylim(0, 0.8)
plt.colorbar(label='Gamma Ray (API)')

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.title('Density Porosity vs Neutron Porosity Scatter Plot', fontsize=14, fontweight='bold')
plt.xlabel('Neutron Porosity (dec)')
plt.ylabel('Density Porosity (dec)')

plt.show()
```

# Changing the point size

```python
plt.figure(figsize=(10,8))
plt.grid(color='lightgray', alpha=0.5, zorder=1)

plt.scatter(df['NPHI'], df['DPHI'], c=df['GR'], vmin=0, vmax=100, zorder=2, s=5, cmap='viridis_r')  # Use of s to control the point size
plt.xlim(0, 0.8)
plt.ylim(0, 0.8)
plt.colorbar(label='Gamma Ray (API)')

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.title('Density Porosity vs Neutron Porosity Scatter Plot', fontsize=14, fontweight='bold')
plt.xlabel('Neutron Porosity (dec)')
plt.ylabel('Density Porosity (dec)')

plt.show()
```

# Annotating Points

```python
# Format : plt.annotate('Text We Want to Display', xy=(x,y), xytext=(x_of_text, y_of_text)
```

```python
plt.figure(figsize=(10,8))
plt.grid(color='lightgray', alpha=0.5, zorder=1)

plt.scatter(df['NPHI'], df['DPHI'], c=df['GR'], vmin=0, vmax=100, cmap='viridis_r',
            zorder=2, s=5)
plt.xlim(0, 0.8)
plt.ylim(0, 0.8)
plt.colorbar(label='Gamma Ray (API)')

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.title('Density Porosity vs Neutron Porosity Scatter Plot', fontsize=14, fontweight='bold')
plt.xlabel('Neutron Porosity (dec)')
plt.ylabel('Density Porosity (dec)')

# For Annotation
plt.scatter(0.42 ,0.17, color='red', marker='o', s=100, zorder=3)  # Use of 100 for s to increase the size of red point
plt.annotate('Shale Point', xy=(0.42 ,0.17), xytext=(0.5, 0.05),
             fontsize=12, fontweight='bold', 
             arrowprops=dict(arrowstyle='->',lw=3), zorder=4)

plt.show()
```

# Highlighting area of interest

```python
# Format for box : plt.gca().add_patch(Rectangle((x_position, y_position), width, height, alpha=0.2, color='yellow'))
# Format for text : plt.text(x_position, y_position, s='Text You Want to Display', fontsize=12, fontweight='bold', ha='center', color='grey')
```

```python
from matplotlib.patches import Rectangle

plt.figure(figsize=(10,8))
plt.grid(color='lightgray', alpha=0.5, zorder=1)

plt.scatter(df['NPHI'], df['DPHI'], c=df['GR'], vmin=0, vmax=100, cmap='viridis_r',
            zorder=2, s=5)
plt.xlim(0, 0.8)
plt.ylim(0, 0.8)
plt.colorbar(label='Gamma Ray (API)')

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.title('Density Porosity vs Neutron Porosity Scatter Plot', fontsize=14, fontweight='bold')
plt.xlabel('Neutron Porosity (dec)')
plt.ylabel('Density Porosity (dec)')

plt.scatter(0.42 ,0.17, color='red', marker='o', s=100, zorder=3)
plt.annotate('Shale Point', xy=(0.42 ,0.17), xytext=(0.5, 0.05),
             fontsize=12, fontweight='bold', 
             arrowprops=dict(arrowstyle='->',lw=3), zorder=4)

plt.text(0.6, 0.75, s='Possible Washout Effects', fontsize=12, fontweight='bold', ha='center', color='grey')
plt.gca().add_patch(Rectangle((0.4, 0.4), 0.4, 0.4, alpha=0.2, color='yellow'))

plt.show()
```
