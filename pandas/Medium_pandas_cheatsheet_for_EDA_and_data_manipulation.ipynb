# 6. Create new features

```python
df['first_letter_code'] = df['Code'].apply(lambda x: x[0])
df['first_letter_code'].value_counts()
```

```python
diz_country = {
    'E': 'England', 
    'S': 'Scotland', 
    'W': 'Wales', 
    'N': 'Northen Ireland',
    'K': 'Kent'
}
df['Country'] = df['first_letter_code'].map(lambda x: diz_country[x])
df['Country'].value_counts(normalize=True)
```

# 7. Sort the dataset based on the value of a feature

```python
df_pop_eng = df.loc[
    (df.Country.isin(['England', 'Wales'])) & (df.Geography == 'Unitary Authority'), 
    ['Code', 'Name', 'Mid-2021 Pop', 'Country']
].sort_values(by='Mid-2021 Pop', ascending=False)
df_pop_eng.head(15)
```

# 8. Group data based on features

```python
df_group_countries = df[
    df.Geography.isin(
        ['Unitary Authority', 'Non-metropolitan District', 'Metropolitan District',
         'London Borough', 'Council Area', 'Local Government District']
    )
].groupby('Country').agg(
    {'Mid-2011 Pop': 'sum', 'Mid-2021 Pop': 'sum'}
).reset_index().sort_values(by='Mid-2021 Pop', ascending=False)

df_group_countries
```

# 9. Add row to Pandas Dataframe

```python
data_row = {
    "Code": [np.nan],
    "Name": ["Channel Islands"],
    "Geography": ["Crown Dependency"],
    "Mid-2021 Pop": [172683],
    "Mid-2011 Pop": [157819],
    "Mid-2001 Pop": [146044],
    "Country": [np.nan]
}
df_channelislands = pd.DataFrame.from_dict(data_row) 
df_channelislands
```

```python
df_c = pd.concat(
    [df, df_channelislands],
    ignore_index=True
)
df_c.tail()
```

# 10. Join two dataframes

```python
df_households = pd.read_csv("data/TS041-2021-1.csv")
df_households
```

```python
df_merge = df_c.merge(
    df_households,
    how='left',
    left_on='Code',
    right_on='Lower Tier Local Authorities Code'
)
df_merge.head()
```

```python
df_merge.isnull().sum()
```

# Summary

- `head()` and `tail()` to visualize the first and last five rows of the dataset.
- `describe()` to calculate the statistics of all the columns.
- `unique()` and `nunique()` to obtain, respectively, the unique values of a column and the number of unique values.
- `value_counts()` to see the frequencies of each modality of your categorical field.
- `isnull().sum()` to check the missing values of all the columns.
- `apply()` and `map()` functions to create new features.
- `groupby()` function to group data based on one or more features.
- `concat()` function to add one or more rows to a dataframe.
- `merge()` function to join two dataframes.
