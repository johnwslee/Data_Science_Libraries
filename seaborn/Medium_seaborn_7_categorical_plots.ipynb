# Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

# Data Read In

```python
df = pd.read_csv("data/diabetes_data_upload.csv")
df
```

```python
# Setting RC parameters (figure size)
rc_params = {"figure.dpi": 120, "savefig.dpi": 300}
sns.set(rc_params)
```

# 1. Paired Bar Plot

```python
sns.barplot(data=df,
            x="Polyuria",
            y="Age",
            hue="Gender",
            palette="husl")
plt.title("Polyuria Analysis Using Bar Plot")

# Place legend outside top right corner of plot
plt.legend(bbox_to_anchor=(1.02, 1), loc="upper left", borderaxespad=0)
plt.show();
```

# 2. Count Plot

```python
sns.countplot(data=df,
              y="Gender",
              hue="partial paresis",
              palette="flare")
plt.title("Count Plot for Partial Paresis")
plt.show();
```

# 3. Strip Plot

```python
sns.stripplot(data=df,
              x="Age",
              y="visual blurring",
              hue="Gender",
              palette="cubehelix")
plt.title("Multivariant Analysis")
plt.show();
```

# 4. Swarm Plot

```python
sns.catplot(data=df,
            x="weakness",
            y="Age",
            hue="Gender",
            kind="swarm",  # "strip", "swarm", "box", "violin", "boxen", "point", "bar", or "count"
            palette="Set2")
plt.title("Swarm Plot for Weakness")
plt.show();
```

# 5. Box Plot

```python
sns.catplot(data=df,
            x="Obesity",
            y="Age",
            hue="Gender",
            kind="box",  # "strip", "swarm", "box", "violin", "boxen", "point", "bar", or "count"
            palette="pastel")
plt.title("Box Plot for Obesity")
plt.show();
```

# 6. Violin Plot

```python
sns.catplot(data=df,
            x="delayed healing",
            y="Age",
            hue="Gender",
            kind="violin",  # "strip", "swarm", "box", "violin", "boxen", "point", "bar", or "count"
            palette="light:#5A9")
plt.title("Violin Plot for Delayed Healing")
plt.show();
```

```python
sns.violinplot(data=df,
               x="delayed healing",
               y="Age",
               hue="Gender",
               palette="light:#5A9")
plt.title("Violin Plot for Delayed Healing")
plt.show();
```

# 7. Distribution Plot

```python
sns.displot(data=df,
            x="Age",
            hue="Gender",
            col="Gender",
            palette="husl")
plt.title("Distibution Plot for Age")
plt.show();
```
