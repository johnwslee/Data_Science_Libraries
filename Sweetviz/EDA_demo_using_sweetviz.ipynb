# Imports

```python
import sweetviz as sw
import pandas as pd

from sklearn.model_selection import train_test_split
```

# Data Read-In

```python
data = pd.read_csv("titanic.csv")
```

```python
data
```

# Generation of EDA Report

```python
# analyzing the data
analyze_report = sw.analyze(data)
```

```python
# Displaying the analyzed data in the browser
analyze_report.show_html('output.html', open_browser=True)  # if open_browser=False, it generate output.html file instead
```

```python
# Displaying the analyzed data in the notebook
analyze_report.show_notebook()
```

# Data Splitting into Train/Test

```python
# Separating X and y

X = data.drop(columns=['Survived'])
y = data['Survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

# EDA on train and test data

```python
# analyzing the train/test data and compare it
comparison_report = sw.compare([X_train, 'Train'], [X_test, 'Test'])
```

```python
# Displaying the analyzed data in the browser
comparison_report.show_html('output.html', open_browser=True)  # if open_browser=False, it generate output.html file instead
```

```python
# Displaying the analyzed data in the notebook
comparison_report.show_notebook()
```
